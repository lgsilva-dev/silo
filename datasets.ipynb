```python
encoded_dairy['Production Date'] = pd.to_datetime(encoded_dairy['Production Date'])
encoded_dairy['Expiration Date'] = pd.to_datetime(encoded_dairy['Expiration Date'])
encoded_dairy['Date'] = pd.to_datetime(encoded_dairy['Date'])


encoded_dairy['Days Until Expiration'] = (encoded_dairy['Expiration Date'] - encoded_dairy['Production Date']).dt.days
encoded_dairy['Days Since Production'] = (encoded_dairy['Date'] - encoded_dairy['Production Date']).dt.days
encoded_dairy['Days to Sell'] = (encoded_dairy['Expiration Date'] - encoded_dairy['Date']).dt.days
```

```python
df_clean = encoded_dairy.copy()
```

```python
df_clean = df_clean.drop(columns=['Date', 'Production Date', 'Expiration Date'])
```

```python
# Identify numerical columns manually
interval_cols = [
    'Total Land Area (acres)',
    'Number of Cows',
    'Quantity (liters/kg)',
    'Price per Unit',
    'Total Value',
    'Shelf Life (days)',
    'Quantity Sold (liters/kg)',
    'Price per Unit (sold)',
    'Approx. Total Revenue(INR)',
    'Quantity in Stock (liters/kg)',
    'Minimum Stock Threshold (liters/kg)',
    'Reorder Quantity (liters/kg)',
    'Days Until Expiration',
    'Days Since Production',
    'Days to Sell'
]
```

```python
phik_matrix = df_clean.phik_matrix(interval_cols=interval_cols)
```

```python
phik_matrix
```

#### Matriz de Correlação

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# (Opcional) Remover valores 1.0 da diagonal principal
phik_matrix_clean = phik_matrix.copy()
np.fill_diagonal(phik_matrix_clean.values, np.nan)

# Tamanho da figura ajustado dinamicamente
plt.figure(figsize=(16, 12))

# Gerar o heatmap
sns.heatmap(
    phik_matrix_clean,
    cmap="coolwarm",
    annot=True,
    fmt=".2f",
    linewidths=0.5,
    cbar_kws={"label": "PhiK Correlation"},
    square=True
)

# Layout e título
plt.xticks(rotation=90)
plt.yticks(rotation=0)
plt.title("PhiK Correlation Matrix Heatmap", fontsize=18, pad=20)
plt.tight_layout()
plt.show()
```

```python
# Correlações > que 0.5 forte
import numpy as np

high_corr_pairs = (
    phik_matrix.where(np.triu(np.ones(phik_matrix.shape), k=1).astype(bool))
    .stack()
    .reset_index()
    .rename(columns={0: 'phik_correlation', 'level_0': 'Var1', 'level_1': 'Var2'})
    .sort_values(by='phik_correlation', ascending=False)
)

high_corr_pairs.query("phik_correlation > 0.5")
```

```python
import phik
from phik.report import plot_correlation_matrix
from phik import resources
```

```python
import pandas as pd

# Attempt to parse with multiple formats
def safe_to_datetime(col):
    return pd.to_datetime(col, errors='coerce')  # Converts invalid values to NaT

#delhivery['cutoff_timestamp'] = safe_to_datetime(delhivery['cutoff_timestamp'])
delhivery['od_start_time'] = safe_to_datetime(delhivery['od_start_time'])
delhivery['od_end_time'] = safe_to_datetime(delhivery['od_end_time'])
delhivery = delhivery.drop(["cutoff_timestamp"], axis = 1)

# Check how many values failed conversion
print(delhivery[['od_start_time', 'od_end_time']].isna().sum())
```

```python
# Convert timestamps to numerical values (seconds since epoch)
delhivery['trip_creation_time'] = pd.to_datetime(delhivery['trip_creation_time'])
delhivery['trip_creation_time'] = delhivery['trip_creation_time'].astype(int) / 10**9  # Convert to seconds
```

```python
print(delhivery[delhivery['cutoff_timestamp'].isna()].head())
```

```python
# Convert timestamps to datetime if not already
delhivery['od_start_time'] = pd.to_datetime(delhivery['od_start_time'])
delhivery['od_end_time'] = pd.to_datetime(delhivery['od_end_time'])

# Feature Engineering: Extract time-based features
delhivery['od_start_hour'] = delhivery['od_start_time'].dt.hour
delhivery['od_start_dayofweek'] = delhivery['od_start_time'].dt.dayofweek
delhivery['od_start_elapsed'] = (delhivery['od_start_time'] - delhivery['od_start_time'].min()).dt.total_seconds()

delhivery['od_end_hour'] = delhivery['od_end_time'].dt.hour
delhivery['od_end_dayofweek'] = delhivery['od_end_time'].dt.dayofweek
delhivery['od_end_elapsed'] = (delhivery['od_end_time'] - delhivery['od_end_time'].min()).dt.total_seconds()

# Drop the original high-cardinality datetime columns
delhivery.drop(columns=['od_start_time', 'od_end_time'], inplace=True)
```

```python
import pandas as pd
import phik


# Drop high-cardinality categorical columns
high_card_cols = ['destination_center', 'destination_name', 'route_schedule_uuid', 'source_center','source_name','trip_uuid']
delhivery_reduced = delhivery.drop(columns=high_card_cols)

# Manually define interval (continuous numeric) columns
# Updated interval column list with engineered features
interval_cols = [
    'start_scan_to_end_scan', 'cutoff_factor', 'actual_distance_to_destination',
    'actual_time', 'osrm_time', 'osrm_distance', 'factor',
    'segment_actual_time', 'segment_osrm_time', 'segment_osrm_distance', 'segment_factor',
    'od_start_hour', 'od_start_dayofweek', 'od_start_elapsed',
    'od_end_hour', 'od_end_dayofweek', 'od_end_elapsed'
]


# Compute Phik correlation matrix with explicit interval columns
phik_matrix = delhivery_reduced.phik_matrix(interval_cols=interval_cols)

# Display result
print(phik_matrix)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# (Opcional) Remover valores 1.0 da diagonal principal
phik_matrix_clean = phik_matrix.copy()
np.fill_diagonal(phik_matrix_clean.values, np.nan)

# Tamanho da figura ajustado dinamicamente
plt.figure(figsize=(16, 12))

# Gerar o heatmap
sns.heatmap(
    phik_matrix_clean,
    cmap="coolwarm",
    annot=True,
    fmt=".2f",
    linewidths=0.5,
    cbar_kws={"label": "PhiK Correlation"},
    square=True
)

# Layout e título
plt.xticks(rotation=90)
plt.yticks(rotation=0)
plt.title("PhiK Correlation Matrix Heatmap", fontsize=18, pad=20)
plt.tight_layout()
plt.show()
```
